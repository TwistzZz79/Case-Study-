## Logistic Classification

```python
from sklearn.model_selection import train_test_split
```

```python
df_full_train, df_test = train_test_split(df, test_size=0.2, random_state=1)
df_train, df_val = train_test_split(df_full_train, test_size=0.25, random_state=1)
```

```python
len(df_train), len(df_val), len(df_test)
```

```python
df_train = df_train.reset_index(drop=True)
df_val = df_val.reset_index(drop=True)
df_test = df_test.reset_index(drop=True)
```

```python
y_train = df_train.delay.values
y_val = df_val.delay.values
y_test = df_test.delay.values

del df_train['delay']
del df_val['delay']
del df_test['delay']
```

```python
df_full_train = df_full_train.reset_index(drop=True)
```

```python
categories=['season','supplier','po_category','destination/country','qty_ord','warehouse','mode_of_transport']
```

```python
from sklearn.feature_extraction import DictVectorizer
```

```python
dv = DictVectorizer(sparse=False)

train_dict = df_train[categories].to_dict(orient='records')
X_train = dv.fit_transform(train_dict)

val_dict = df_val[categories].to_dict(orient='records')
X_val = dv.transform(val_dict)
```

```python
from sklearn.linear_model import LogisticRegression
```

```python
model = LogisticRegression(max_iter=1000)
model.fit(X_train, y_train)
```

```python
y_pred = model.predict_proba(X_val)[:, 1]
```

```python
delay_decision=y_pred>=0.5
```

```python
df_pred=pd.DataFrame()
df_pred['probability']=y_pred
df_pred['prediction']=delay_decision.astype(int)
df_pred['actual']=y_val
```

```python
df_pred['correct']=df_pred.prediction==df_pred.actual
```
